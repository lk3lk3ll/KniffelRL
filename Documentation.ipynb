ZIEL

In diesem Projekt werde ich versuchen einem Neuronalen Netzwerk mithilfe des REINFORCE algorithmus das Spiel "Kniffel" beizubringen. (https://www.geeksforgeeks.org/machine-learning/reinforce-algorithm/) Dabei benutze ich keinen Diskontsatz, da ich nur einmal meinen Gewinn berechne. Außerdem benutze ich einen normierten Gewinnbetrag um den Alghorithmus skalierungsunabhängig zu machen. Das heißt für jede Episode berechne ich den Mittelwert und die Standartabweichung des Gewinns aus allen Spielen und rechne den Gewinn um: normierter Gewinn = (Gewinn - Mittelwert)/Abweichung.

BIBLIOTHEKEN

Ich verwende das Reinforcement Learning Framework "Gymnasium" (https://gymnasium.farama.org/) um die Umgebung für das Spiel zu realisieren und die Bibliothek "PyTorch" (https://pytorch.org/) für das Trainieren meines neuronalen Netzwerks. Des weiteren benutze ich die Bibliothek "Pickle" (https://docs.python.org/3/library/pickle.html) um die erreichten Gewinne zu speichern und die Bibliothek "MatPlotLib" (https://matplotlib.org/) um diese zu visualisieren.

KNIFFEL ENVIRONMENT

Im Kniffel environment definiere ich zunächst die Würfel, welche eine Augenzahl haben
 ,gewählt sein können (das NN will diesen Würfel erneut würfeln)
 und frei sein können(der Würfel darf nochmal gewürfelt werden).
  Und ich definiere auch die Kombinationen, welche frei sein können (man hat die Möglichkeit
  diese Kombination zu wählen) und eine Punktzahl haben können (wenn man eine Kombination wählt,
   wird die neue Punktzahl in score eingetragen und die Kombination ist nicht mehr frei).


Damit unser Programm mit dem Spiel arbeiten kann benutzen wir One hot Encoding.d.h-___ In der Klasse Kniffel Env definiere ich nun die Liste action_space in der alle Aktionen sind die der Spieler in Kniffel machen kann. Er kann jeden der 5 Würfel zum erneuten Würfeln wählen (Aktion 0-4)
zum nächsten Schritt gehen (Aktion 5) oder eine der 13 kombinationen auswählen (6-18).


Ich definieren auch noch die Liste observation_space bestehenend aus:
- den drei stages in denen der Spieler sein kann: 1.Anfang der Runde, 2. nur ein mal neu gewürfelt, ein mal geht noch 3. zwei mal neu gewürfelt, man muss eine Kombination wählen (one hot encoded)
- aus jeder Augenzahl die jeder Würfel haben kann und den Eigenschaften jenes Würfels also ob der Spieler diesen Würfel gewählt hat und ob man ihn erneut werfen kann (5*8, one hot encoded),
- den zuvor erwähnten 13 Kombinationen
- allen Punkten die nötig sind um den Bonus zu bekommen (0-63)

insgesamt beträgt diese Liste also 120 Elemente.

POLICY

Die Policy wird als "Deep Neural Network" realisiert. Ich verwende die Pyramiden Architektur mit drei fully connected linear hidden layers. Diese verwenden ReLU als Aktivierungsfunktion. Die Ausgangslayer verwendet die Softmax Aktivierungsfunktion.

DATEIEN

Die wichtigsten Dateien sind:
- CalcScore.py beinhaltet Methoden zur Berechnung der Gewinne
- KniffelEnv.py realisiert die Gymnasium Umgebung für Kniffel
- DeepNeuralNetworkPolicy.py realisiert das Neuronale Netzwerk
- PlayGame.py beinhaltet die Methode play_game, welche ein Spiel spielen kann und die Informationen für den REINFORCE Algorithmus zurückgibt: das ist die Spieltraektorie in Form von logorithmischen Warscheinlichkeiten der Aktionen und das Spiel Ergebnis
- TrainPolicy.py beinhaltet den training Algorithmus. Dieser schreibt die Zwischenergebnisse auf die Festplatte und kann unterbrochen und neu gestartet werden. Sie schreibt zwei Dateien: Das neuronale Netz und die erreichten Gewinne für jede Epoche
